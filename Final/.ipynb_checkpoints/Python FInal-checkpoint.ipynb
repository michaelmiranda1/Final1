### Circular Plot of 12 segments corresponding to months of the year with the radius of each segment corresponding to the relative rates of failure for that month:

** Time of the year when more banks tend to fail more

```python
#import plotly.plotly as py
#import plotly.graph_objs as go

#df4 = pd.read_csv('Monthclosuretotals.csv')
#df4.head()

trace1 = go.Barpolar(
    r=[51, '-', '-', 59, '-', '-', 74, '-', '-', 59, '-', '-'],
    text=['Jan', '-', '-', 'Apr', '-', '-', 'Jul', '-', '-', 'Oct', '-', '-'],
    name='+50 Banks in Month',
    marker=dict(
        color='rgb(106,81,163)'
    )
)

trace2 = go.Barpolar(
    r=['-', 44, 38, '-', 43, 36, '-', 41, 38, '-', 35, 35],
    text=['-', 'Feb', 'Mar', '-', 'May', 'Jun', '-', 'Aug', 'Sep', '-', 'Nov', 'Dec'],
    name='-50 Banks in Month',
    marker=dict(
        color='rgb(158,154,200)'
    )
)

data = [trace1, trace2]
layout = go.Layout(
    title='Bank Failures By Month',
    font=dict(
        size=16
    ),
    legend=dict(
        font=dict(
            size=16
        )
    ),
    radialaxis=dict(
        ticksuffix='%'
    ),
    orientation=270
)
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='polar-area-chart')
```

### Cloropleth map showing number of failures by each state

** Compare each state's bank failures

```python
# Chorpleth Without Puerto Rico but with the Hover Feature


import plotly.plotly as py
import pandas as pd

df = pd.read_csv('Statefail.csv')

for col in df.columns:
    df[col] = df[col].astype(str)

scl = [[0.0, 'rgb(242,240,247)'],[0.2, 'rgb(218,218,235)'],[0.4, 'rgb(188,189,220)'],\
            [0.6, 'rgb(158,154,200)'],[0.8, 'rgb(117,107,177)'],[1.0, 'rgb(84,39,143)']]

df['text'] = df['States'] + '<br>' +\
    'Failure Count '+df['Failure_Count']

data = [ dict(
        type='choropleth',
        colorscale = scl,
        autocolorscale = False,
        locations = df['States'],
        z = df['Failure_Count'].astype(float),
        locationmode = 'USA-states',
        text = df['text'],
        marker = dict(
            line = dict (
                color = 'rgb(255,255,255)',
                width = 2
            ) ),
        colorbar = dict(
            title = "Bank Failures")
        ) ]

layout = dict(
        title = 'United States Bank Failure from 2000-2018<br>(Hover for breakdown)',
        geo = dict(
            scope='usa',
            projection=dict( type='albers usa' ),
            showlakes = True,
            lakecolor = 'rgb(255, 255, 255)'),
             )
    
fig = dict( data=data, layout=layout )
py.iplot( fig, filename='d3-cloropleth-map' )
```

```python
#Choropleth with Puerto Rico Without the Hover Feature



import pandas as pd
import folium
 
# Load the shape of the zone (US states)
# Find the original file here: https://github.com/python-visualization/folium/tree/master/examples/data
# You have to download this file and set the directory where you saved it
state_geo = '/Users/michaelmiranda/pycache/Final/us-stateswpr.json'
 
# Load the unemployment value of each state
# Find the original file here: https://github.com/python-visualization/folium/tree/master/examples/data 
state_data = pd.read_csv('Statefail.csv')
 
# Initialize the map:
m = folium.Map(location=[37, -102], zoom_start=5)
 
# Add the color for the chloropleth:
m.choropleth(
 geo_data=state_geo,
 name='choropleth',
 data=state_data,
 columns=['States', 'Failure_Count'],
 key_on='feature.id',
 fill_color='YlGn',
 fill_opacity=0.7,
 line_opacity=0.2,
 legend_name='Failure Count'
)
folium.LayerControl().add_to(m)
 
# Save to html
m.save('StateBank.html')

```


#### Observable pattern to the banks that fail to be acquired, related to date, location, or number of failures in a state?


Observable patterns to banks in relation to date is that the banks that the majority of the banks that are left over are at the beginning of the year. If the fiscal year is in October then that would equate to the end of the 2nd quarter. If the closing of some accounts in some banks is at the beginning of the year then that would equate at the end of the first quarter. Despite all that, banks are left that are left over are predominately at the beginning of the New Year.

Bank closures in general in the first wind rose graph though is very noticable that they are cyclical that they close every quarter. Further more in analyzing the data in the first graph, the bank closures you can take notice that most of the banks closed in wake of the housing market crash.
